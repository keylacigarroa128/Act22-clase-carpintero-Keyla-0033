```python
class Carpintero:
    def __init__(self, nombre, telefono, edad):
        self.nombre = nombre
        self.telefono = telefono
        self.edad = edad

    def construir(self):
        return f"-> {self.nombre} está construyendo un mueble sólido."

    def instalar(self):
        return f"-> {self.nombre} está instalando gabinetes de cocina."

    def reparar(self):
        return f"-> {self.nombre} está reparando un mueble dañado."

def registrar_carpintero(numero):
    print(f"\n--- Registro del Carpintero {numero} ---")
    nombre = input("Nombre: ")
    telefono = input("Teléfono: ")
    edad = input("Edad: ")
    return Carpintero(nombre, telefono, edad)

# --- Interacción con el usuario ---

# Creamos los objetos mediante la interacción
carpintero1 = registrar_carpintero(1)
carpintero2 = registrar_carpintero(2)

# Mostramos resultados
for carp in [carpintero1, carpintero2]:
    print(f"\nAcciones realizadas por {carp.nombre}:")
    print(carp.construir())
    print(carp.instalar())
    print(carp.reparar())


}print("Programa elaborado por Keyla Paola Cigarroa Palacios NL = 12")
```
